```python
import pandas as pd


df = pd.read_excel("SSI_VIC_Financial_Ratio_13112023.xlsx")
```

```python
# Reset and drop Indexes
df = df.iloc[6:]
df = df.iloc[:-4]   #Drop the "powered by fiintrade"
df = df.reset_index(drop=True)
# Tranpose and reset the index again
df = df.transpose().reset_index(drop=True)
# Set the first row as headers
new_headers = df.iloc[0]
df = df[1:]
df.columns = new_headers
# df = df.rename(columns={"ITEMS": "Quarter"})
```

```python
df
```

```python
# Add identifiers
# Assuming df is your DataFrame
cols=pd.Series(df.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [dup + '_' + str(i) if i != 0 else dup for i in range(sum(cols == dup))]
df.columns=cols
```

```python
# Get all the columns name where its not "VIC"
valid_columns = df.columns[~df.columns.astype(str).str.contains('VIC')].to_list()
```

```python
type(valid_columns)
```

```python
len(valid_columns)
```

```python
tickers_columns = df.columns[df.columns.astype(str).str.contains("VIC")].tolist()
```

```python
len(tickers_columns)
```

```python
valid_columns
```

```python
valid_columns = [col for col in valid_columns if col not in ["Ratio", "Profit Growth (%)", "Revenue Growth (%)"]]
```

```python
# Name mapping from tickers to valid names
name_mapping = dict(zip(tickers_columns, valid_columns))

# Remove the empty columns first
df = df.drop(columns = valid_columns)

# Rename 
df.rename(columns=name_mapping, inplace=True)
```
